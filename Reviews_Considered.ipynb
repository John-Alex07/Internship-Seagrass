```python
review_df = df[['review_title', 'review_description', 'variety']]
review_df
```

```python
review_df['review_title'] = review_df['review_title'].apply(str.lower)
review_df['review_description'] = review_df['review_description'].apply(str.lower)
review_df['variety'] = review_df['variety'].apply(str.lower)
```

```python
review_df
```

```python
from nltk.corpus import stopwords
```

```python
stop_words_italy = stopwords.words('italian')
stop_words_german = stopwords.words('german')
stop_words_france = stopwords.words('french')
stop_words_portugal = stopwords.words('portuguese')
stop_words_spain = stopwords.words('spanish')
stop_words = stopwords.words('english')
```

```python
stop_words_portugal
```

```python
from nltk.tokenize import word_tokenize
```

```python
review_df
```

```python
# review_df['review_title'] = review_df['review_title'].str.replace('bianco', 'white')
```

```python
for i in review_df.columns:
    review_df[i] = review_df[i].apply(word_tokenize)
review_df
```

```python
import string
```

```python
stop_punc = list(string.punctuation)
stop_punc
```

```python
def rem_stop(tokens):
    return [word for word in tokens if word not in stop_words and word not in stop_punc and 
            word not in stop_words_france and word not in stop_words_german and word not in stop_words_italy
            and word not in stop_words_portugal and word not in stop_words_spain]
for i in review_df.columns:
    review_df[i] = review_df[i].apply(rem_stop)
review_df
```

```python
for i in review_df.columns:
    review_df[i] = review_df[i].apply(' '.join)
review_df
```

```python
import re
```

```python
review_df['review_title'] = review_df['review_title'].apply(lambda x : re.sub(r'\d+', '', x))
review_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
cvectorizer = CountVectorizer()
```

```python
X = review_df.iloc[:, 0]
Y = review_df.iloc[:, -1]
```

```python
data = X
```

```python
data
```

```python
data = cvectorizer.fit_transform(data)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data,Y,test_size=0.3, random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
```

```python
lg.fit(X_train, Y_train)
```

```python
lg.score(X_test, Y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb = MultinomialNB()
```

```python
nb.fit(X_train, Y_train)
```

```python
nb.score(X_test, Y_test)
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
nbb = BernoulliNB()
```

```python
nbb.fit(X_train, Y_train)
```

```python
nbb.score(X_test, Y_test)
```
